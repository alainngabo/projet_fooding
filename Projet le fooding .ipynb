6) Manipuler ce json (dictionnaire) pour y extraire des informations sur les restaurants :
'type', 'name', 'address', 'phone_num', 'cellphone_num', 'cravings', 'extras',
'price','sorted_tags',’latitude’,longitude ‘,...

```python
for elem in d['results']:
    print(elem.keys())
    print('typerestaurant:', elem['type'])
    typerestaurant.append(elem['type'])
    print('nom:', elem['name'])
    nom.append(elem['name'])
    print('Ville:', elem['address']['city'])
    ville.append(elem['address']['city'])
    print('code postal:', elem['address']['postal_code'])
    codepostal.append(elem['address']['postal_code'])
    print('Rue:', elem['address']['street_name_1'])
    rue.append(elem['address']['street_name_1'])
    print('Rue 2:', elem['address']['street_name_2'])
    rue2.append(elem['address']['street_name_2'])
    print('Téléphone:', elem['phone_num'])
    telephone.append(elem['phone_num'])
    print ('Mobile:', elem['cellphone_num'])
    mobile.append(elem['cellphone_num'])
    print ('Cravings:', '.' .join(elem['cravings']))
    cravings.append(',  '.join(elem['cravings']))
    print ('Extras:', '.' .join(elem['extras']))
    extras.append(',  '.join(elem['extras']))               
    print ('Prix:','.' .join(elem['price']))
    prix.append(',  '.join(elem['price']))             
    print ('Tags: station de métro:','.' .join(elem['sorted_tags']['transport']))
    stationdemetro.append(',  '.join(elem['sorted_tags']['transport']))
    print('Latitude',elem['address']['position'][0])
    print('Longitude',elem['address']['position'][1])
    Latitude.append(elem['address']['position'][0])
    Longitude.append(elem['address']['position'][1])
    print('###################')
```

```python
#Create a map
carte = folium.Map(location=[48.85,2.35],tiles='openstreetmap',zoom_start=13)

#Creation of clusters and subgroup
mcg = MarkerCluster(control=False)
carte.add_child(mcg)

a = folium.plugins.FeatureGroupSubGroup(mcg,'ville')
carte.add_child(a)

#remplissage de la carte
for idx, row in df.iterrows():
    Marker([row['Latitude'], row['Longitude']],popup= row['nom']+" Fourchette des prix: "+row['prix']+"  Spécialités:  "+row['cravings'],
          icon=folium.Icon(color="red", icon="info-sign")).add_to(a)
   

folium.LayerControl(collapsed=True).add_to(carte)
carte.save('CarteDDL.html')
carte
```
